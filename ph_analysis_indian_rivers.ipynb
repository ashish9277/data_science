```python
#code to check PH levels of several indian rivers over the years 2001-2011

# the data is taken from Indian Government Website 
#https://data.gov.in/resources/water-quality-few-indian-rivers-2001-2011/download

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline

river_data = pd.read_csv('River_Data.csv')

ph_river_data = river_data[river_data['Parameter'] == 'pH']

#fetching the rivers
rivers = ph_river_data['Category'].tolist()

rivers = [i.split(',')[0] for i in rivers] 

#the following can be fetched directly from dataset head. Need to work on this further...!!!
years = [2001,2002,2003,2004,2005,2006,2007,2008]

#the ph values are defined in terms of range. So while plotting we are calculating the mean of the values
upper = float(ph_river_data['2001'].tolist()[0].split('-')[0])
lower = float(ph_river_data['2001'].tolist()[0].split('-')[1])

#following is a raw code.. Variables can be reiterated from lists
ph_2001_data = river_data[river_data['Parameter'] == 'pH']['2001'].tolist()
ph_2002_data = river_data[river_data['Parameter'] == 'pH']['2002'].tolist()
ph_2003_data = river_data[river_data['Parameter'] == 'pH']['2003'].tolist()
ph_2004_data = river_data[river_data['Parameter'] == 'pH']['2004'].tolist()
ph_2005_data = river_data[river_data['Parameter'] == 'pH']['2005'].tolist()
ph_2006_data = river_data[river_data['Parameter'] == 'pH']['2006'].tolist()
ph_2007_data = river_data[river_data['Parameter'] == 'pH']['2007'].tolist()
ph_2008_data = river_data[river_data['Parameter'] == 'pH']['2008'].tolist()
ph_2009_data = river_data[river_data['Parameter'] == 'pH']['2009'].tolist()
ph_2010_data = river_data[river_data['Parameter'] == 'pH']['2010'].tolist()
ph_2011_data = river_data[river_data['Parameter'] == 'pH']['2011'].tolist()

ph_2001_numbers = []
for i in ph_2001_data:
    upper = float(i.split('-')[0])
    lower = float(i.split('-')[1])
    ph_2001_numbers.append(np.mean([upper,lower]))
    
ph_2002_numbers = []
for i in ph_2002_data:
    upper = float(i.split('-')[0])
    lower = float(i.split('-')[1])
    ph_2002_numbers.append(np.mean([upper,lower]))
    
ph_2003_numbers = []
for i in ph_2003_data:
    upper = float(i.split('-')[0])
    lower = float(i.split('-')[1])
    ph_2003_numbers.append(np.mean([upper,lower]))
    
ph_2004_numbers = []
for i in ph_2004_data:
    upper = float(i.split('-')[0])
    lower = float(i.split('-')[1])
    ph_2004_numbers.append(np.mean([upper,lower]))
    
ph_2005_numbers = []
for i in ph_2005_data:
    upper = float(i.split('-')[0])
    lower = float(i.split('-')[1])
    ph_2005_numbers.append(np.mean([upper,lower]))
    
ph_2006_numbers = []
for i in ph_2005_data:
    upper = float(i.split('-')[0])
    lower = float(i.split('-')[1])
    ph_2006_numbers.append(np.mean([upper,lower]))

ph_2007_numbers = []
for i in ph_2005_data:
    upper = float(i.split('-')[0])
    lower = float(i.split('-')[1])
    ph_2007_numbers.append(np.mean([upper,lower]))
    
ph_2008_numbers = []
for i in ph_2005_data:
    upper = float(i.split('-')[0])
    lower = float(i.split('-')[1])
    ph_2008_numbers.append(np.mean([upper,lower]))
    
ph_2009_numbers = []
for i in ph_2005_data:
    upper = float(i.split('-')[0])
    lower = float(i.split('-')[1])
    ph_2009_numbers.append(np.mean([upper,lower]))
    
ph_2010_numbers = []
for i in ph_2005_data:
    upper = float(i.split('-')[0])
    lower = float(i.split('-')[1])
    ph_2010_numbers.append(np.mean([upper,lower]))
    
ph_2011_numbers = []
for i in ph_2005_data:
    upper = float(i.split('-')[0])
    lower = float(i.split('-')[1])
    ph_2011_numbers.append(np.mean([upper,lower]))
    

# plotting the graph
sns.set(style="whitegrid")
sns.set_color_codes("pastel")

f, ax = plt.subplots(figsize=(10, 6))

plt.xticks(years,rivers)

plt.plot(years, ph_2001_numbers, label='2001', color='r')
plt.plot(years, ph_2002_numbers, label='2002', color='g')
plt.plot(years, ph_2003_numbers, label='2003', color='b')
plt.plot(years, ph_2004_numbers, label='2004', color='orange')
plt.plot(years, ph_2005_numbers, label='2005', color='black')
plt.plot(years, ph_2006_numbers, label='2006', color='brown')
plt.plot(years, ph_2007_numbers, label='2008', color='gray')
plt.plot(years, ph_2008_numbers, label='2008', color='violet')
plt.plot(years, ph_2009_numbers, label='2009', color='c')
plt.plot(years, ph_2010_numbers, label='2010', color='m')
plt.plot(years, ph_2011_numbers, label='2011', color='chartreuse')

#the limits can also be set from low to high based on min and max
plt.xlim(2001, 2008)
ax.set_ylabel('pH Value')
ax.set_xlabel('River')
ax.set_title('Average ph value over years')
legend = plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=True, borderpad=1, borderaxespad=1)
```
